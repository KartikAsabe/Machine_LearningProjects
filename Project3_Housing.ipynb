```python
import numpy as np
import pandas as pd
```

```python
data= pd.read_csv('/content/Housing.csv')
```

```python
data.head()
```

```python
data.tail()
```

```python
#Onehot coding
data['guestroom']=data['guestroom'].apply({'yes':1, 'no':2}.get)
data['basement']=data['basement'].apply({'yes':1, 'no':2}.get)
data['hotwaterheating']=data['hotwaterheating'].apply({'yes':1, 'no':2}.get)
data['airconditioning']=data['airconditioning'].apply({'yes':1, 'no':2}.get)
data['prefarea']=data['prefarea'].apply({'yes':1, 'no':2}.get)
data['furnishingstatus']=data['furnishingstatus'].apply({'furnished':1, 'semi-furnished':2, 'unfurnished':3}.get)
data['mainroad']=data['mainroad'].apply({'yes':1, 'no':2}.get)
```

```python
data.head()
```

```python
data.head()
```

```python
#Dividing the data into depndent and independent variable
x = data[['area','bedrooms','bathrooms','stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea','furnishingstatus']]
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8)
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
predictions=regression.predict(x_test)
```

```python
regression.score(x,y)
```

Ridge Regression

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=1)
ridge.fit(x_train,y_train)
```

```python
predictions_ridge=ridge.predict(x_test)
```
