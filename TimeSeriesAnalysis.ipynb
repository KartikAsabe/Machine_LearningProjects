```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
from statsmodels.tsa.ar_model import AutoReg
data = [10,12,14,16,18,20,20,22,24,26]
```

```python
plt.plot(data)
```

```python
ar_model=AutoReg(data, lags=3)
ar_model_fit=ar_model.fit()
```

```python
#forecasting for 10,11,12 (Three values)
forecast=ar_model_fit.predict(10,12)
print(forecast)
```

```python
forecast_list=forecast.tolist() #array gets coverted to list
```

```python
my_data= data+forecast_list
```

```python
print(my_data)
```

```python
plt.plot(my_data)
```
